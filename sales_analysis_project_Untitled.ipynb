### Data quality Observation and Cleaning-
we see null rows in list of orders, we see trailing spaces,different date formats and negative numbers in profit columns in order details table.
Prompt- wrote a code for data cleaning by removing fully empty rows,strip spaces from the possible joining keys,convert all dates to standard format in the three dataframes. Create 'month' field in the List of orders dataframe using Order Date  field to be used as joining key with Sales Target data. Suggest any other feature engineering requirements for performing merge and EDA in the next steps.

```python
# ==========================================================
# DATA QUALITY CLEANING PIPELINE
# For: List of Orders, Order Details, Sales Target
# ==========================================================

# -----------------------------
# 1. REMOVE FULLY EMPTY ROWS
# Issue: trailing blank rows in List of Orders
# -----------------------------
list_of_order_df = list_of_order_df.dropna(how="all")

# -----------------------------
# 2. STANDARDIZE JOIN KEYS
# Issue: hidden spaces / inconsistent types in Order ID
# -----------------------------
list_of_order_df["Order ID"] = list_of_order_df["Order ID"].astype(str).str.strip()
orders_details_df["Order ID"] = orders_details_df["Order ID"].astype(str).str.strip()

# -----------------------------
# 3. CONVERT ORDER DATE (LIST OF ORDERS)
# Issue: date stored as string (DD-MM-YYYY)
# -----------------------------
list_of_order_df["Order Date"] = pd.to_datetime(
    list_of_order_df["Order Date"],
    format="%d-%m-%Y",
    errors="coerce"
)

# -----------------------------
# 4. CONVERT SALES TARGET MONTH COLUMN
# Issue: month stored as string (Apr-18)
# -----------------------------
sales_target_df["Month of Order Date"] = pd.to_datetime(
    sales_target_df["Month of Order Date"],
    format="%b-%y",
    errors="coerce"
)

# -----------------------------
# 5. CONVERT NUMERIC COLUMNS SAFELY
# Issue: numeric columns may be object type or corrupted
# -----------------------------
for col in ["Amount", "Profit", "Quantity"]:
    orders_details_df[col] = pd.to_numeric(
        orders_details_df[col],
        errors="coerce"
    )

sales_target_df["Target"] = pd.to_numeric(
    sales_target_df["Target"],
    errors="coerce"
)

# -----------------------------
# 6. CONVERT CATEGORICAL COLUMNS
# Issue: repeated string columns optimized to category type
# -----------------------------
cat_cols_orders = ["CustomerName", "State", "City"]
for col in cat_cols_orders:
    list_of_order_df[col] = (
        list_of_order_df[col]
        .astype(str)
        .str.strip()
        .astype("category")
    )

cat_cols_details = ["Category", "Sub-Category"]
for col in cat_cols_details:
    orders_details_df[col] = (
        orders_details_df[col]
        .astype(str)
        .str.strip()
        .astype("category")
    )

sales_target_df["Category"] = (
    sales_target_df["Category"]
    .astype(str)
    .str.strip()
    .astype("category")
)

# -----------------------------
# 7. HANDLE MISSING VALUES SAFELY
# Issue: NaNs after conversions (especially dates/numerics)
# -----------------------------
# Option: drop or keep depending on analysis need
list_of_order_df = list_of_order_df.dropna(subset=["Order ID", "Order Date"])
orders_details_df = orders_details_df.dropna(subset=["Order ID"])

# -----------------------------
# 8. CREATE TIME FEATURES (FOR FUTURE EDA)
# -----------------------------
list_of_order_df["Order Month"] = list_of_order_df["Order Date"].dt.to_period("M").dt.to_timestamp()

# -----------------------------
# 9. VALIDATE DATA QUALITY
# -----------------------------
print("DATA QUALITY CHECK SUMMARY")
print("=" * 40)

print("\nMissing values per dataset:")
print("Orders:\n", list_of_order_df.isnull().sum())
print("\nOrder Details:\n", orders_details_df.isnull().sum())
print("\nSales Target:\n", sales_target_df.isnull().sum())

print("\nShapes after cleaning:")
print("Orders:", list_of_order_df.shape)
print("Order Details:", orders_details_df.shape)
print("Sales Target:", sales_target_df.shape)

print("\nDuplicate Order IDs (Orders):", list_of_order_df["Order ID"].duplicated().sum())
print("Duplicate Order IDs (Details):", orders_details_df["Order ID"].duplicated().sum())

# -----------------------------
# 10. READY FOR MERGE CONFIRMATION
# -----------------------------
print("\nData cleaning completed. Ready for merging and EDA.")
```

### Advance Data Cleaning 
Prompt- Ensure joining keys are unique and not missing. This will help in avoiding any data loss/rows drops during merge and keep the total consistent.
Prompt- scenarios excluding where entire row is empty, look for the possible unique keys or joining keys and if that value is missing, replace missing or blank customer_id values with stable UNKNOWN_ IDs, and print how many IDs were filled. Ensure to follow the join structure of- 
List of Orders ↔ Order Details → Order ID
Sales Target ↔ Category + Month
So focus on:
Order ID
Category
Month of Order Date

```python
# ==========================================================
# FUNCTION: FIX MISSING JOIN KEYS SAFELY
# ==========================================================

def fix_missing_keys(df, key_cols, df_name="DataFrame"):
    """
    Replaces missing/blank join keys with stable UNKNOWN_ IDs
    and prints how many values were fixed.
    """

    print("=" * 70)
    print(f"KEY CLEANING REPORT: {df_name}")
    print("=" * 70)

    for col in key_cols:

        if col not in df.columns:
            print(f"[SKIP] Column '{col}' not found")
            continue

        # Standardize blanks → NaN
        df[col] = df[col].astype(str).str.strip()
        df[col] = df[col].replace(["", "nan", "None", "NaN"], pd.NA)

        # Count missing before fill
        missing_count = df[col].isna().sum()

        # Fill missing with stable IDs
        if missing_count > 0:
            if col == "Order ID":
                df[col] = df[col].fillna("UNKNOWN_ORDER_ID")
            elif col == "Category":
                df[col] = df[col].fillna("UNKNOWN_CATEGORY")
            elif col == "Month of Order Date":
                df[col] = df[col].fillna("UNKNOWN_MONTH")
            else:
                df[col] = df[col].fillna(f"UNKNOWN_{col.upper()}")

        print(f"{col}: filled {missing_count} missing values")

    print("\n")
    return df
```

```python

# -----------------------------
# 1. List of Orders
# -----------------------------
list_of_order_df = fix_missing_keys(
    list_of_order_df,
    key_cols=["Order ID"],
    df_name="List of Orders"
)

# -----------------------------
# 2. Order Details
# -----------------------------
orders_details_df = fix_missing_keys(
    orders_details_df,
    key_cols=["Order ID", "Category"],
    df_name="Order Details"
)

# -----------------------------
# 3. Sales Target
# -----------------------------
sales_target_df = fix_missing_keys(
    sales_target_df,
    key_cols=["Category", "Month of Order Date"],
    df_name="Sales Target"
)
```

### Perform OUTER MERGE on three datasets and check the data integrity with audit of the final dataset
Prompt- perform outer merge on the list of orders and order details dataset with following conditions-
keeps all records= outer join
validate-one to many
flags origin using indicator=True
create a readable merge status column and help detect dropped/mismatched rows with an audit report checking all required columns, count of rows and missing/mismatches values

```python

# ==========================================================
# OUTER MERGE WITH FULL AUDIT TRAIL
# ==========================================================

sales_df = pd.merge(
    list_of_order_df,
    orders_details_df,
    on="Order ID",
    how="outer",
    indicator=True,
    validate="one_to_many"   # correct relationship
)

# ----------------------------------------------------------
# Create readable merge status column
# ----------------------------------------------------------

sales_df["merge_status"] = sales_df["_merge"].map({
    "both": "Matched in both datasets",
    "left_only": "Only in Orders (missing details)",
    "right_only": "Only in Order Details (orphan rows)"
})

# ----------------------------------------------------------
# Quick summary of merge quality
# ----------------------------------------------------------

print("=" * 60)
print("MERGE QUALITY REPORT")
print("=" * 60)

print("\nRow counts by source:")
print(sales_df["_merge"].value_counts())

print("\nTotal rows after merge:", sales_df.shape[0])

print("\nMerge status breakdown:")
print(sales_df["merge_status"].value_counts())
```

```python
#Filter clean dataset for analysis
clean_sales_df = sales_df[sales_df["_merge"] == "both"].copy()
#Drop helper column
clean_sales_df.drop(columns=["_merge"], inplace=True)
```

#### MERGE WITH SALES TARGET (OUTER JOIN + AUDIT)
prompt-join the sales_df (orders + order details) dataset with the sales_target_df (Month + Category targets) using following criteria- outer join (no data loss hidden)
indicator=True
create a readable merge status column and help detect dropped/mismatched rows with an audit report checking all required columns, count of rows and missing/mismatches values

```python

# ==========================================================
# MERGE WITH SALES TARGET (OUTER JOIN + AUDIT)
# ==========================================================

sales_target_merged = pd.merge(
    sales_df,
    sales_target_df,
    left_on=["Order Month", "Category"],
    right_on=["Month of Order Date", "Category"],
    how="outer",
    indicator="target_merge_flag",
    validate="many_to_many"
)

# ----------------------------------------------------------
# Create readable merge status
# ----------------------------------------------------------

sales_target_merged["merge_status"] = (
    sales_target_merged["target_merge_flag"]
    .map({
        "both": "Matched in Actual + Target",
        "left_only": "Actual only (no target match)",
        "right_only": "Target only (no actual sales)"
    })
)

# ----------------------------------------------------------
# Summary report
# ----------------------------------------------------------

print("=" * 70)
print("SALES vs TARGET MERGE QUALITY REPORT")
print("=" * 70)

print("\nRow distribution:")
print(sales_target_merged["target_merge_flag"].value_counts())

print("\nTotal rows after merge:")
print(sales_target_merged.shape[0])

print("\nStatus breakdown:")
print(sales_target_merged["merge_status"].value_counts())
```

### Create the HTML version of the graphs.
prompt-Give me a python code to create a html version of the graphs in a dashboard style for easy view.

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Create output directory
os.makedirs("eda_html_output", exist_ok=True)

# ==========================================================
# GRAPH 1 — MONTHLY SALES TREND
# ==========================================================

monthly_sales = (
    sales_df
    .groupby("Order Month", as_index=False)["Amount"]
    .sum()
)

plt.figure(figsize=(12,6))

sns.lineplot(
    data=monthly_sales,
    x="Order Month",
    y="Amount",
    marker="o",
    linewidth=2.5,
    color="royalblue"
)

plt.title("Monthly Sales Trend", weight="bold")
plt.xlabel("Order Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("eda_html_output/monthly_sales.png", dpi=300)
plt.close()

# ==========================================================
# GRAPH 2 — CATEGORY SALES
# ==========================================================

category_sales = (
    sales_df
    .groupby("Category", as_index=False)["Amount"]
    .sum()
    .sort_values("Amount", ascending=False)
)

plt.figure(figsize=(12,6))

sns.barplot(
    data=category_sales,
    x="Category",
    y="Amount",
    palette="viridis"
)

plt.title("Category-wise Total Sales", weight="bold")
plt.tight_layout()

plt.savefig("eda_html_output/category_sales.png", dpi=300)
plt.close()

# ==========================================================
# GRAPH 3 — CATEGORY PROFITABILITY
# ==========================================================

category_profit = (
    sales_df
    .groupby("Category", as_index=False)["Profit"]
    .sum()
    .sort_values("Profit", ascending=False)
)

plt.figure(figsize=(12,6))

sns.barplot(
    data=category_profit,
    x="Category",
    y="Profit",
    palette="coolwarm"
)

plt.axhline(0, color="black")
plt.title("Category-wise Profit Analysis", weight="bold")
plt.tight_layout()

plt.savefig("eda_html_output/category_profit.png", dpi=300)
plt.close()

# ==========================================================
# GRAPH 4 — TARGET VS ACTUAL
# ==========================================================

target_vs_actual = (
    sales_target_merged
    .groupby("Category", as_index=False)
    .agg({
        "Amount": "sum",
        "Target": "sum"
    })
)

target_vs_actual = target_vs_actual.melt(
    id_vars="Category",
    value_vars=["Amount", "Target"],
    var_name="Metric",
    value_name="Value"
)

plt.figure(figsize=(12,6))

sns.barplot(
    data=target_vs_actual,
    x="Category",
    y="Value",
    hue="Metric",
    palette="Set2"
)

plt.title("Actual Sales vs Sales Target", weight="bold")
plt.tight_layout()

plt.savefig("eda_html_output/target_vs_actual.png", dpi=300)
plt.close()

```
